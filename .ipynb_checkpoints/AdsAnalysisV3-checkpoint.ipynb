# import the BeautifulSoup function from bs4 to strip the ads of html tags

```python
from bs4 import BeautifulSoup

def cleanhtml2(raw_html):
    cleantext = BeautifulSoup(raw_html, "lxml").text
    return cleantext
```

```python
dfTot['clean_cont'] = dfTot['content'].apply(lambda x: cleanhtml2(x))
```

# function for cleaning text (can probably use a package function to do this..., NLTK? Spacy?)
> this function does the trick for cleaning text of the ads and converting to lower case

```python
# function for text cleaning
def clean_text(text):
    # remove backslash-apostrophe
    text = re.sub("\'", "", text)
    # remove everything alphabets
    text = re.sub("[^a-zA-Z]"," ",text)
    # remove whitespaces
    text = ' '.join(text.split())
    # convert text to lowercase
    text = text.lower()
    
    return text
```

```python
dfTot['clean_cont'] = dfTot['clean_cont'].apply(lambda x: clean_text(x))
```

# get rid of stopwords using NLTK

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
#stop_words = set(stopwords.words('english'))
stop_words = nltk.corpus.stopwords.words('english')
newStopWords = ['edu','com','au','experience','skills','development','work','role','team']
stop_words.extend(newStopWords)

# function to remove stopwords
def remove_stopwords(text):
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)
  
dfTot['clean_cont'] = dfTot['clean_cont'].apply(lambda x: remove_stopwords(x))
```

```python
dfTot['clean_cont'][0]
```

```python
def freq_words(x, terms = 30):
  all_words = ' '.join([text for text in x])
  all_words = all_words.split()
  
  fdist = nltk.FreqDist(all_words)
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())})
  
  # selecting top 20 most frequent words
  d = words_df.nlargest(columns="count", n = terms) 
  plt.figure(figsize=(12,15))
  ax = sns.barplot(data=d, x= "count", y = "word")
  ax.set(ylabel = 'Word')
  plt.show()
```

```python
# print 100 most frequent words
freq_words(dfTot['clean_cont'], 100)
```

# Investigating terms highly-correlated with job classifications

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(sublinear_tf=True, min_df=10, norm='l2', encoding='latin-1', ngram_range=(1, 3), stop_words='english', max_features=10000)

features = tfidf.fit_transform(dfTot.clean_cont).toarray()
#labels = dfTot[dfTot.class_id==range(29)].class_id
labels = dfTot.class_id
features.shape
```

# testing and exploring the process that I will use to tag in what follows 
> based on TFIDF terms from the entire corpus that are most correlated with particular labels (here the labels are the job classifications i.e. 30 categories 

> here we explore the most highly-correlated unigrams, bigrams and trigrams with particular categories (trigrams generally appear to be considerably less meaningful (or at least relevant to job skills and responsibilities) than uni and bigrams. 

> later in the notebook (in the section titled "training several ML classifiers for job classification output, given ad content features", I also begin to look at unigrams/bi- and trigrams that are 'important' for each job category based upon their contributions to the performance of a set of ML classifiers.

```python
from sklearn.feature_selection import chi2
N = 10
for classification_name, class_id in sorted(class_to_id.items()):
  features_chi2 = chi2(features, labels == class_id)
  indices = np.argsort(features_chi2[0])
  feature_names = np.array(tfidf.get_feature_names())[indices]
  unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
  bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
  trigrams = [v for v in feature_names if len(v.split(' ')) == 3]
  print("# '{}':".format(classification_name))
  print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
  print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-N:])))
  print("  . Most correlated trigrams:\n. {}".format('\n. '.join(trigrams[-N:])))
```

# Now use the same procedure for each job ad
> take the uni-, bi-, trigrams from a particular job ad and find which of them are to be found 
in the 'top 100 (or whatever number is chosen)' most highly correlated TFIDF terms for the particular label that the document belongs to

> use the 'CountVectorizer' class to process each ad into its words/wordsets to match with the TFIDF set

# in the below, the parameter 'numDocs' indicates the number of ads (starting at 0 in the dataframe dfTot), that will be annotated 


```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(encoding='latin-1', ngram_range=(1, 3), stop_words='english', max_features=1000)

listTags = []
numDocs = 50

for iDoc in range(numDocs):
    features_chi2 = chi2(features, labels == dfTot.class_id[iDoc])
    indices = np.argsort(features_chi2[0])

    features1 = count_vect.fit([dfTot.clean_cont[iDoc]])
    #features1 = tfidf.fit_transform(dfTot.clean_cont).toarray()
    #labels = dfTot[dfTot.class_id==range(29)].class_id
    labels1 = dfTot.class_id[iDoc]

    list1 = np.array(count_vect.get_feature_names())
    list2 = np.array(tfidf.get_feature_names())[indices]
    tags=[i for i in list1 if i in list2[-100:]] 
    #note that the argument '100' in this line can be increased 
    #to increase the likelihood of matching a term in the document with one in the most highly-correlated TFIDF terms
    #
    listTags.append(tags)
    print(iDoc)
```

# training several ML classifiers for job classification output, given ad content features

> the procedure we use below to tag ads is essentially identical to that used above (where we only consider overlap of uni-, bi-, tri-grams in specific job ads with the TFIDF terms that are highly-correlated with job classification labels

> here, however, we use coefficient values of a Linear Support Vector Machine model to denote the 'importance' of each of the features (i.e. uni-, bi-, tri-grams and therefore whether they should be included in the annotation tags for the ad

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

X = dfTot['clean_cont'] # Collection of documents
y = dfTot['classification_name'] # Target or the labels we want to predict (i.e., the 13 different complaints of products)

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25,
                                                    random_state = 0)
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

clf = MultinomialNB().fit(X_train_tfidf, y_train)
```

```python
print(clf.predict(count_vect.transform(["microsoft"])))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC

from sklearn.model_selection import cross_val_score


models = [
    RandomForestClassifier(n_estimators=200, max_depth=3, random_state=0),
    LinearSVC(),
    MultinomialNB(),
    LogisticRegression(random_state=0),
]
CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
  model_name = model.__class__.__name__
  accuracies = cross_val_score(model, features, labels, scoring='accuracy', cv=CV)
  for fold_idx, accuracy in enumerate(accuracies):
    entries.append((model_name, fold_idx, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])
```

```python
import seaborn as sns

sns.boxplot(x='model_name', y='accuracy', data=cv_df)
sns.stripplot(x='model_name', y='accuracy', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=2)
plt.show()
```

```python
cv_df.groupby('model_name').accuracy.mean()
```
